```python
# test one hot encoding for k nearest classification

%%script echo pass

categoricals = ['anaemia', 'diabetes', 'high_blood_pressure', 'sex', 'smoking']
for column in categoricals:
  encode = pd.get_dummies(df[column])
  if column == 'sex':
    # as per female = 0, male = 1 in canvas discussions
    encode.columns = ['female', 'male']
  else:
    encode.columns = [column+'False', column+'True']
  df.drop(column, axis=1, inplace=True)
  df = df.join(encode)

df
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import StratifiedKFold
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import export_text
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.utils import shuffle
from sklearn.model_selection import GridSearchCV
```

```python
# Base data models

# %%script echo pass

X = df.drop('DEATH_EVENT', axis=1)
y = df.DEATH_EVENT
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

dtree = DecisionTreeClassifier().fit(X_train, y_train)
dtreepred = dtree.predict(X_test)
print(f"Decision tree results: [{dtree.score(X_train, y_train):.2f}, {dtree.score(X_test, y_test):.2f}]")
print(confusion_matrix(y_test, dtreepred))
print(classification_report(y_test, dtreepred))

knn = KNeighborsClassifier().fit(X_train, y_train)
knnpred = knn.predict(X_test)
print(f"k nearest results: [{knn.score(X_train, y_train):.2f}, {knn.score(X_test, y_test):.2f}]")
print(confusion_matrix(y_test, knnpred))
print(classification_report(y_test, knnpred))
```

```python
# Figure 20
# Text representation of the base decision tree

print(export_text(dtree, feature_names=list(X.columns)))
```

```python
# Decision tree feature selection

# %%script echo pass

cols = []
maxScore = 0.0
randomCol = shuffle(range(0, df.shape[1]-1))

print("Test set accuracy with features selected:")
for column in range(0, df.shape[1]-1):
  cols.append(randomCol[column])
  newData = df.drop('DEATH_EVENT', axis=1).values[:, cols]
  X_train, X_test, y_train, y_test = train_test_split(newData, y, test_size=0.2)
  X_train = scaler.fit_transform(X_train)
  X_test = scaler.fit_transform(X_test)
  clf = DecisionTreeClassifier()
  fit = clf.fit(X_train, y_train)
  score = clf.score(X_test, y_test)
  if score < maxScore:
    cols.remove(randomCol[column])
  else:
    maxScore = score
    print(f"{df.columns[cols[-1]]}: {score}")

dX = df.iloc[:, cols]
dy = df.DEATH_EVENT
dX_train, dX_test, dy_train, dy_test = train_test_split(dX, dy, test_size=0.2)
scaler = MinMaxScaler()
X_train = scaler.fit_transform(dX_train)
X_test = scaler.transform(dX_test)

dtree = DecisionTreeClassifier().fit(dX_train, dy_train)
dtreepred = dtree.predict(dX_test)
print(f"Decision tree feature selection results: [{dtree.score(dX_train, dy_train):.2f}, {dtree.score(dX_test, dy_test):.2f}]")
print(confusion_matrix(dy_test, dtreepred))
print(classification_report(dy_test, dtreepred))
```

```python
# k nearest feature selection

%%script echo pass

cols = []
maxScore = 0.0
randomCol = shuffle(range(0, df.shape[1]-1))

print("Test set accuracy with features selected:")
for column in range(0, df.shape[1]-1):
  cols.append(randomCol[column])
  newData = df.drop('DEATH_EVENT', axis=1).values[:, cols]
  X_train, X_test, y_train, y_test = train_test_split(newData, y, test_size=0.2)
  X_train = scaler.fit_transform(X_train)
  X_test = scaler.fit_transform(X_test)
  clf = KNeighborsClassifier()
  fit = clf.fit(X_train, y_train)
  score = clf.score(X_test, y_test)
  if score < maxScore:
    cols.remove(randomCol[column])
  else:
    maxScore = score
    print(f"{df.columns[cols[-1]]}: {score}")

kX = df.iloc[:, cols]
ky = df.DEATH_EVENT
kX_train, kX_test, ky_train, ky_test = train_test_split(kX, ky, test_size=0.2)
scaler = MinMaxScaler()
kX_train = scaler.fit_transform(kX_train)
kX_test = scaler.transform(kX_test)

knn = KNeighborsClassifier().fit(kX_train, ky_train)
knnpred = knn.predict(kX_test)
print(f"k nearest feature selection results: [{knn.score(kX_train, ky_train):.2f}, {knn.score(kX_test, ky_test):.2f}]")
print(confusion_matrix(ky_test, knnpred))
print(classification_report(ky_test, knnpred))
```

```python
# Parameter hypertuning using stratified kfolds cross validation

# %%script echo pass

max_depth = list(range(2, 15))
min_samples_leaf = list(range(5,50))
criterion = ["gini", "entropy"]
leaf_size = list(range(1,50))
n_neighbors = list(range(1,30))
p = [1,2]
dparameters = dict(max_depth=max_depth, min_samples_leaf=min_samples_leaf, criterion=criterion)
kparameters = dict(leaf_size=leaf_size, n_neighbors=n_neighbors, p=p)
skf = StratifiedKFold(n_splits=5)

dtree = DecisionTreeClassifier()
knn = KNeighborsClassifier()
dtuner = GridSearchCV(dtree, dparameters, cv=skf, refit=True).fit(X,y)
ktuner = GridSearchCV(knn, kparameters, cv=skf, refit=True).fit(X,y)

print("Hypertuned decision tree parameters:")
print(dtuner.best_estimator_)
hypertree = dtuner.best_estimator_.fit(dX_train, dy_train)
print(f"Hypertuned training set accuracy: {hypertree.score(dX_train, dy_train):.2f}")
print(f"Hypertuned test set accuracy: {hypertree.score(dX_test, dy_test):.2f}")
print()
print("Hypertuned k nearest parameters:")
print(ktuner.best_estimator_)
hyperknn = ktuner.best_estimator_.fit(kX_train, ky_train)
print(f"Hypertuned training set accuracy: {hyperknn.score(kX_train, ky_train):.2f}")
print(f"Hypertuned test set accuracy: {hyperknn.score(kX_test, ky_test):.2f}")
print()
print("Stratified kfolds cross validation hypertuned feature selection decision tree model:")
dtreepred = dtuner.predict(dX_test)
print(confusion_matrix(dy_test, dtreepred))
print(classification_report(dy_test, dtreepred))
print("Stratified kfolds cross validation hypertuned feature selection k nearest model:")
knnpred = ktuner.predict(kX_test)
print(confusion_matrix(ky_test, knnpred))
print(classification_report(ky_test, knnpred))
```
